```python
X=df.iloc[:,[3,4]].values
```

```python
X
```

# choose the cluster number
    WSS=within cluster some of squres


```python
#wcss values for different number of cluster
wcss=[]
for i in range(1,11):
 kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
 kmeans.fit(X)
    
 wcss.append(kmeans.inertia_)
```

```python
#plot an elbo graph
sns.set()
plt.plot(range(1,11),wcss)
plt.title('the elbo plot graph')
plt.xlabel('Nember of cluster')
plt.ylabel('wcss')
plt.show()
```

# tranning the k means clustri g model
 Optimum number of cluster =5

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
 # return a label for each data point  base on their cluster
Y=kmeans.fit_predict(X)
print(Y)
```

# Visualize all the cluster

```python
#plot all the cluster
plt.figure(figsize=(8,8))
plt.scatter(X[Y==0,0],X[Y==0,1],s=50,c='green',label='cluster 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=50,c='red',label='cluster 2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=50,c='yellow',label='cluster 3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=50,c='violet',label='cluster 4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=50,c='blue',label='cluster 5')
            
#plot the center
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label='Centroids')
plt.title('customer group')
plt.xlabel('Annual income')
plt.ylabel('Spending Score')
plt.show()
```
